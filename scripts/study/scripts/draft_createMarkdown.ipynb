```python
import resourcesdb
import pandas as pd
import pprint
import collections
import createMarkdown
```

```python
rf = pd.DataFrame(resourcesdb.getResourcesDict()).T
```

```python
mc = rf.main_category.unique()
mc.sort()
print(mc)
```

```python
tables = []
for c in mc:
    a = rf[(rf.main_category == c) & (rf.studying_now == False)]
    if a.empty:
        continue
    tables.append(a)
    
```

```python
b = tables[1]
print(b)
```

```python
c = b.T.to_dict()
c = collections.defaultdict(dict, c)
pprint.pprint(c)
```

```python
props = []
for i in tables:
    a = i.T.to_dict()
    a = collections.defaultdict(dict, a)
    pprint.pprint(a)
    props.append(a)
```

```python
pprint.pprint(props[1])
```

```python
text = ''
for i in props:
    cat = (i[list(i.keys())[0]]['main_category'])
    text += createMarkdown.tableHeader(textBefore=f"\n## {cat}\n\n")
    text += createMarkdown.tableRow(i)
    text += '\n\n'
    
print(text)
    
    
```
